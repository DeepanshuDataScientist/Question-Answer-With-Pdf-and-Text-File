```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings.sentence_transformer import (
    SentenceTransformerEmbeddings,
)
from langchain_text_splitters import CharacterTextSplitter
```

```python
loader = TextLoader("/content/My_Resume.txt")
documents = loader.load()
documents
```

```python
# split into chunks

text_spliter = CharacterTextSplitter(chunk_size = 1000, chunk_overlap = 0)
docs = text_spliter.split_documents(documents)
docs
```

```python
len(docs)
```

```python
!pip install sentence-transformers
```

```python
# create the open-source embedding function
embedding_function = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

embedding_function
```

```python
# load it into Chroma
db = Chroma.from_documents(docs, embedding_function)
db

# query it
query = "What are the contents in Resume"
docs = db.similarity_search(query)

# print results
print(docs[0].page_content)
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("/content/DeepanshuResume.pdf")
pages = loader.load_and_split()
```

```python
pages[0]
```

```python
# split into chunks

text_spliter = CharacterTextSplitter(chunk_size = 1000, chunk_overlap = 0)
docs = text_spliter.split_documents(pages)
docs
```

```python
len(docs)
```

```python
!pip install sentence-transformers
```

```python
# create the open-source embedding function
embedding_function = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

embedding_function
```

```python
# load it into Chroma
db = Chroma.from_documents(docs, embedding_function)
# db

# query it
query = "Please provide the the content in resume"
docs = db.similarity_search(query)


# print results
print(docs[0].page_content)

```
